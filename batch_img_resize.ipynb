```python
from PIL import Image
import os

# Input and output folder paths
input_folder = "./MyBlackDataset/black"  # Replace with the path to your folder
output_folder = "./MyBlackDataset/black_resized"  # Replace with the path to save resized images

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Loop through all files in the input folder
for filename in os.listdir(input_folder):
    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, filename)

    try:
        # Open an image file
        with Image.open(input_path) as img:
            # Resize the image to 768x768
            resized_img = img.resize((768, 768))
            # Save it to the output folder
            resized_img.save(output_path)
            print(f"Resized and saved: {filename}")
    except Exception as e:
        print(f"Failed to process {filename}: {e}")
```
